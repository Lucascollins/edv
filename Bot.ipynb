```python
#Importando bibliotecas
from selenium import webdriver
import tkinter as tk
import pandas as pd
import pyautogui
import time


#Layout
root = tk.Tk()
root.geometry('370x400')
root.resizable(width=0, height=0)
root.title("ADM Sender")
root["bg"] ="#1C402A"

#Labels
_nameSheet= Label(root,text ="Nome Planilha",bg="#1C402A",fg="white",font=("Poppins Boldd",15))
_nameSheet.grid(column = 0,row = 0,padx=50,pady=10)
_inputSheet = Entry(root,width = 50,bg="#89FE87")
_inputSheet.grid(column = 0,row = 1,pady=10,padx=40)

text_mensagem = Label(root,text ="Mensagem",bg="#1C402A",fg="white",font=("Poppins Bold",15))
text_mensagem.grid(column = 0,row = 2,padx=50)
text = Entry(root,width =50,bg="#89FE87")
text.grid(column = 0,row = 3,pady=10,padx=40)


_imageText = Label(root,text="Copie a imagem com ctrl + c para enviar no botÃ£o abaixo",bg="#1C402A",fg="white",font =("Poppins",10))
_imageText.grid(column = 0,row=6)

#Logica do programa
            
def excel():
    n = 0
    _sheet = _inputSheet.get()
    _text = text.get()
    df =pd.read_excel(f"C:/Users/start/Downloads/{_sheet}.xlsx")
    alunos = df["Aluno(a)"]
    telefones = df["Telefone(s)"]
    num = len(telefones)
    
    nav = webdriver.Chrome()
    nav.get('https://web.whatsapp.com/')

    while len(nav.find_elements_by_id('side')) < 1:
        time.sleep(1)

    while n <= len(telefones): 
        link =f"https://web.whatsapp.com/send?phone={telefones[n]}&text={_text}"
        nav.get(link)
        time.sleep(45)
        botao_enviar = nav.find_element_by_xpath(
                "//span[@data-icon='send']")
        botao_enviar.click()
        
        n+= 1

def image_sender():
    n = 0
    _sheet = _inputSheet.get()
    df =pd.read_excel(f"C:/Users/start/Downloads/{_sheet}.xlsx")
    alunos = df["Aluno(a)"]
    telefones = df["Telefone(s)"]
    num = len(telefones)
    nav = webdriver.Chrome()
    nav.get('https://web.whatsapp.com/')

    while len(nav.find_elements_by_id('side')) < 1:
        time.sleep(1)

    while n <= len(telefones): 
        link =f"https://web.whatsapp.com/send?phone={telefones[n]}"
        nav.get(link)
        time.sleep(45)
        pyautogui.hotkey("ctrl","v")
        botao_enviar = nav.find_element_by_xpath(
                "//span[@data-icon='send']")
        botao_enviar.click()
        n+= 1


#Buttons
button_message = Button(root,bg="#89FE87",
                text="Send Message",
                fg="#1C402A",
                font=("Poppins Bold",10),
                command=excel)
button_message.grid(column = 0 ,row=5,pady=20)
#Second
button_image = Button(root,bg="#89FE87",
                text="Send Image",
                fg="#1C402A",
                font=("Poppins Bold",10),
                command=image_sender)
button_image.grid(column = 0 ,row=7)

 


root.mainloop()
```
